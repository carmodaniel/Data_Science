```python
# Definir o caminho da pasta onde estão os arquivos CSV
import os
import pandas as pd

# Caminho da pasta
pasta = '/content/drive/MyDrive/Data_Science/planilha_2025'

# Listar os arquivos CSV dentro da pasta
arquivos = [f for f in os.listdir(pasta) if f.endswith('.csv')]
print("Arquivos encontrados:", arquivos)
```

```python
# Ler todos os CSVs e unir em um único DataFrame
# Lista para armazenar os DataFrames
dfs = []

for arquivo in arquivos:
    caminho_completo = os.path.join(pasta, arquivo)
    print(f"Lendo: {arquivo}")

    # Ajuste encoding e separador se necessário
    df_temp = pd.read_csv(caminho_completo, sep=';', quotechar='"')

    dfs.append(df_temp)

# Concatenar todos em um único DataFrame
df = pd.concat(dfs, ignore_index=True)

print("\nTodos os arquivos foram lidos e unidos com sucesso!")
print("Dimensões do dataset final:", df.shape)
```

```python
# Copiar o dataframe (para evitar mudanças no original)
dados = df.copy()

# Converter colunas numéricas com vírgulas em float
cols_valores = [
    'a_vencer_ate_90_dias', 'a_vencer_de_91_ate_360_dias',
    'a_vencer_de_361_ate_1080_dias', 'a_vencer_de_1081_ate_1800_dias',
    'a_vencer_de_1801_ate_5400_dias', 'a_vencer_acima_de_5400_dias',
    'vencido_acima_de_15_dias', 'carteira_ativa',
    'carteira_inadimplida_arrastada', 'ativo_problematico'
]

for c in cols_valores:
    dados[c] = dados[c].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
```

```python
# GRÁFICO 1 – Distribuição da carteira por UF (Estados)
# Este gráfico mostra onde o volume de crédito está concentrado no país.
# Assim, conseguimos entender a atuação regional e potenciais riscos concentrados.

uf_agrupado = dados.groupby('uf')['carteira_ativa'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=uf_agrupado.index, y=uf_agrupado.values, palette='viridis')
plt.title('Top 10 Estados com Maior Carteira Ativa (R$)', fontsize=14)
plt.ylabel('Valor Total (R$)')
plt.xlabel('UF')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Este gráfico mostra os estados com maior volume de crédito ativo.")
print("- Ele ajuda a identificar regiões com maior exposição financeira e oportunidade de expansão.\n")
```

```python
# GRÁFICO 2 – Modalidades de crédito mais utilizadas
# ==========================================
# Aqui identificamos quais tipos de produtos têm maior representatividade na carteira.
# Isso indica o perfil dos clientes e os tipos de operação mais rentáveis.

mod_agrupado = dados.groupby('modalidade')['carteira_ativa'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=mod_agrupado.index, x=mod_agrupado.values, palette='crest')
plt.title('Top 10 Modalidades de Crédito por Valor de Carteira Ativa', fontsize=14)
plt.xlabel('Valor Total (R$)')
plt.ylabel('Modalidade')
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Identifica os produtos com maior volume de crédito.")
print("- Útil para direcionar esforços comerciais e gestão de risco por tipo de operação.\n")
```

```python
# GRÁFICO 3 – Comparação entre Carteira Ativa e Ativos Problemáticos
# ==========================================
# Mostra a relação entre o total de crédito ativo e os ativos com problemas (inadimplência).
# É um indicador direto da qualidade da carteira.

resumo = dados[['carteira_ativa', 'ativo_problematico']].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=resumo.index, y=resumo.values, palette=['#4CAF50','#E74C3C'])
plt.title('Comparativo: Carteira Ativa x Ativos Problemáticos', fontsize=14)
plt.ylabel('Valor Total (R$)')
plt.xlabel('')
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Compara o total de crédito ativo com o volume de ativos problemáticos.")
print("- Um valor alto em 'Ativos Problemáticos' indica risco elevado e necessidade de medidas de cobrança.\n")
```

```python
# GRÁFICO 4 – Inadimplência por porte do cliente
# ==========================================
# Mostra o percentual de inadimplência (carteira inadimplida / carteira ativa)
# para cada porte, permitindo entender onde está o maior risco.

inad_por_porte = dados.groupby('porte')[['carteira_ativa','carteira_inadimplida_arrastada']].sum()
inad_por_porte['percentual_inadimplencia'] = (inad_por_porte['carteira_inadimplida_arrastada'] / inad_por_porte['carteira_ativa']) * 100
inad_por_porte = inad_por_porte.sort_values('percentual_inadimplencia', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=inad_por_porte.index, y=inad_por_porte['percentual_inadimplencia'], palette='rocket')
plt.title('Percentual de Inadimplência por Porte do Cliente', fontsize=14)
plt.ylabel('% Inadimplência')
plt.xlabel('Porte do Cliente')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Mostra quais faixas de clientes têm maior taxa de inadimplência.")
print("- Ajuda a direcionar políticas de crédito mais rigorosas para segmentos de maior risco.\n")
```

```python
# GRÁFICO 5 – Distribuição dos valores por tempo de vencimento
# ==========================================
# Permite entender o perfil da carteira (curto, médio e longo prazo).
# É essencial para gestão de liquidez e previsão de risco.

prazo_cols = [
    'a_vencer_ate_90_dias', 'a_vencer_de_91_ate_360_dias',
    'a_vencer_de_361_ate_1080_dias', 'a_vencer_de_1081_ate_1800_dias',
    'a_vencer_de_1801_ate_5400_dias', 'a_vencer_acima_de_5400_dias'
]
valores_prazo = dados[prazo_cols].sum()

plt.figure(figsize=(10,6))
sns.barplot(x=valores_prazo.index, y=valores_prazo.values, palette='coolwarm')
plt.title('Distribuição dos Valores a Vencer por Faixa de Prazo', fontsize=14)
plt.ylabel('Valor Total (R$)')
plt.xlabel('Faixa de Prazo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Mostra a concentração do crédito por prazo de vencimento.")
print("- Permite visualizar o equilíbrio entre curto e longo prazo, importante para fluxo de caixa e gestão de risco.\n")
```

```python

# GRÁFICO 6 – Carteira Ativa x Inadimplida por UF
# identificar relações fortes entre as variáveis (ex: carteira ativa x inadimplência).

uf_group = dados.groupby('uf')[['carteira_ativa', 'carteira_inadimplida_arrastada']].sum().nlargest(10, 'carteira_ativa')

uf_group.plot(
    kind='bar',
    stacked=True,
    figsize=(10,6),
    color=['#4CAF50', '#E74C3C']
)
plt.title('Carteira Ativa e Inadimplida por UF (Top 10)', fontsize=14)
plt.xlabel('UF')
plt.ylabel('Valor Total (R$)')
plt.xticks(rotation=45)
plt.legend(['Carteira Ativa', 'Inadimplida'])
plt.tight_layout()
plt.show()

print("Interpretação:")
print("- Mostra simultaneamente o volume de crédito e o total inadimplido por estado.")
print("- Ajuda a identificar regiões com alto risco de crédito ou oportunidades de recuperação.")
print("- Ideal para gestores regionais ou análises de performance por área geográfica.")
```

```python
# Modelo de Classificação

# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

sns.set(style="whitegrid", palette="crest")
plt.rcParams["figure.figsize"] = (8,5)


# Preparação dos dados
dados = df.copy()

# Converter colunas numéricas com vírgula para float
cols_valores = [
    'a_vencer_ate_90_dias', 'a_vencer_de_91_ate_360_dias',
    'a_vencer_de_361_ate_1080_dias', 'a_vencer_de_1081_ate_1800_dias',
    'a_vencer_de_1801_ate_5400_dias', 'a_vencer_acima_de_5400_dias',
    'vencido_acima_de_15_dias', 'carteira_ativa',
    'carteira_inadimplida_arrastada', 'ativo_problematico'
]

for c in cols_valores:
    dados[c] = dados[c].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)

# Remover possíveis valores nulos
dados = dados.dropna()

# Criação da variável alvo (y)
# Queremos prever se o ativo é problemático (risco alto) ou não (risco baixo).
# Risco alto quando o valor de ativo_problematico for maior que 0.

dados['risco_alto'] = np.where(dados['ativo_problematico'] > 0, 1, 0)

print("Distribuição das classes (0=baixo risco, 1=alto risco):")
print(dados['risco_alto'].value_counts(normalize=True))

# Selecionar as features (x) e variável alvo (y)
# Usar as variáveis financeiras como preditoras.
X = dados[[
    'a_vencer_ate_90_dias', 'a_vencer_de_91_ate_360_dias',
    'a_vencer_de_361_ate_1080_dias', 'a_vencer_de_1081_ate_1800_dias',
    'a_vencer_de_1801_ate_5400_dias', 'a_vencer_acima_de_5400_dias',
    'vencido_acima_de_15_dias', 'carteira_ativa', 'carteira_inadimplida_arrastada'
]]
y = dados['risco_alto']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Treinar o modelo Árvore de Decisão
modelo = DecisionTreeClassifier(
    max_depth=4,       # limita a profundidade da árvore para evitar overfitting
    random_state=42
)
modelo.fit(X_train, y_train)

# Avaliar o modelos
y_pred_train = modelo.predict(X_train)
y_pred_test = modelo.predict(X_test)

print("\nAcurácia no Treino:", round(accuracy_score(y_train, y_pred_train), 4))
print("Acurácia no Teste:", round(accuracy_score(y_test, y_pred_test), 4))

print("\n📊 Relatório de Classificação (Teste):")
print(classification_report(y_test, y_pred_test))

# Matriz de confusão
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão - Modelo de Classificação')
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.show()

# Visualizar a árvore
plt.figure(figsize=(20,10))
plot_tree(modelo, feature_names=X.columns, class_names=['Baixo Risco', 'Alto Risco'], filled=True, rounded=True, fontsize=10)
plt.title('Árvore de Decisão - Classificação de Risco de Crédito')
plt.show()

# Interpretação
print("Interpretação:")
print("- O modelo de classificação analisa as variáveis financeiras de cada operação e identifica padrões associados a risco.")
print("- Ele aprendeu, por exemplo, que clientes com altos valores vencidos e ativos problemáticos têm maior probabilidade de risco.")
print("- A Árvore de Decisão permite visualizar claramente as regras usadas para classificar o risco de cada cliente.")
```
